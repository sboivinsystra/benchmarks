```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix, csc_matrix
from scipy.sparse.csgraph import dijkstra
import time
import matplotlib.pyplot as plt
```

```python
links = pd.read_csv('inputs/road_links.csv')
links = links.set_index('index')

edges = links[['a','b','time']].values # build the edges again, useless
sparse, index = sparse_matrix(edges)

np.random.seed(0)
indices = np.random.choice(list(index.values()), 1000, replace=False)
```

```python
def plot_results(x,y_mat,title=''):
    fig, ax = plt.subplots(figsize=(10, 6))
    for i in range(len(y_mat)):
        yy = [1/(i/y[0]) for i in y_mat[i]]
        #plt.plot(x,yy,color=f"{i/10}")
        plt.plot(x,yy,color='red',alpha=(i+1)/11)
    mean = np.mean(y_mat,axis=0)
    yy = [1/(i/y[0]) for i in mean]
    plt.plot(x,yy,color='tab:blue')

    plt.title(title)
    plt.ylabel('x time faster')
    plt.xlabel('num threads')
    t = [f'exec {i+1}' for i in range(len(y_mat))]
    t.append('average')
    plt.legend(t)
    plt.grid(True, 'major', linestyle='-', axis='both')
    ax.set_axisbelow(True)
```

```python
#from Quetzal pathfinder_utils

# buildindex
def build_index(edges):
    nodelist = {e[0] for e in edges}.union({e[1] for e in edges})
    nlen = len(nodelist)
    return dict(zip(nodelist, range(nlen)))

# build matrix
def sparse_matrix(edges, index=None):
    if index is None:
        index = build_index(edges)
    nlen = len(index)
    coefficients = zip(*((index[u], index[v], w) for u, v, w in edges))
    row, col, data = coefficients
    return csr_matrix((data, (row, col)), shape=(nlen, nlen)), index
```

```python
#from Quetzal pathfinder_utils
from concurrent.futures import ProcessPoolExecutor

    
# Wrapper to split the indices (destination) into parallel batchs and compute the shortest path on each batchs.
def ProcessPoolExecutor_dijkstra(csgraph,indices=None,return_predecessors=True, num_core=1,**kwargs):
    '''
    num_core = 1 : number of threads.
    '''
    if num_core == 1:
        return  dijkstra(csgraph=csgraph, indices=indices, return_predecessors=return_predecessors,**kwargs)

    batch = round(len(indices)/num_core)
    indices_mat = [indices[i*batch:(1+i)*batch] for i in range(num_core-1)]
    indices_mat.append(indices[(num_core-1)*batch:])

    results = []
    with ProcessPoolExecutor(max_workers=num_core) as executor:
        for i in range(num_core):
            p = executor.submit(
                dijkstra,
                indices=indices_mat[i],
                csgraph=csgraph,
                return_predecessors=return_predecessors,
                **kwargs
            )
            results.append(p)

    if return_predecessors == True: # result is a tuple
        dist_matrix = np.concatenate([res.result()[0] for res in results],axis=0)
        predecessors = np.concatenate([res.result()[1] for res in results],axis=0).astype(np.int32)
        return dist_matrix, predecessors
    else:
        dist_matrix = np.concatenate([res.result() for res in results],axis=0)
        return dist_matrix
```

```python
#from Quetzal pathfinder_utils
from multiprocessing import Process, Manager


  
# Wrapper to split the indices (destination) into parallel batchs and compute the shortest path on each batchs.
def multiprocessing_dijkstra(csgraph,indices=None,return_predecessors=True, num_core=1,keep_running=True,**kwargs):
    '''
    num_core = 1 : number of threads.
    keep_running = False : if you want to keep the subprocesses alive (mmutiple dijktra in a loop for example)
    '''
    if num_core == 1:
        return  dijkstra(csgraph=csgraph, indices=indices, return_predecessors=return_predecessors,**kwargs)

    batch = round(len(indices)/num_core)
    indices_mat = [indices[i*batch:(1+i)*batch] for i in range(num_core-1)]
    indices_mat.append(indices[(num_core-1)*batch:])
    def process_wrapper(indices, kwargs, result_list, index):
        result = dijkstra(indices=indices, **kwargs)
        result_list[index] = result

    manager = Manager()
    result_list = manager.list([None] * len(indices_mat))
    processes = []

    process_kwargs = {'csgraph':csgraph,'return_predecessors':return_predecessors,**kwargs}
    for i, origins in enumerate(indices_mat):
        process = Process(target=process_wrapper, args=(origins, process_kwargs, result_list, i))
        process.start()
        processes.append(process)
    for process in processes:
        process.join()
    
    result_list = np.array(result_list,dtype="object")

    if return_predecessors == True: # result is a tuple
        dist_matrix = [res[0] for res in result_list]
        dist_matrix = np.concatenate(dist_matrix,axis=0)
        predecessors = [res[1] for res in result_list]
        predecessors = np.concatenate(predecessors,axis=0).astype(np.int32)
        return dist_matrix, predecessors
    else:
        dist_matrix = np.concatenate(result_list,axis=0)
        return dist_matrix
```

```python
# Wrapper to split the indices (destination) into parallel batchs and compute the shortest path on each batchs.
def multiprocessing_dijkstra_no_return(csgraph,indices=None,return_predecessors=True, num_core=1,keep_running=True,**kwargs):
    '''
    num_core = 1 : number of threads.
    keep_running = False : if you want to keep the subprocesses alive (mmutiple dijktra in a loop for example)
    '''
    if num_core == 1:
        return  dijkstra(csgraph=csgraph, indices=indices, return_predecessors=return_predecessors,**kwargs)

    batch = round(len(indices)/num_core)
    indices_mat = [indices[i*batch:(1+i)*batch] for i in range(num_core-1)]
    indices_mat.append(indices[(num_core-1)*batch:])
    def process_wrapper(indices, kwargs, result_list, index):
        result = dijkstra(indices=indices, **kwargs)
        result_list[index] = result

    manager = Manager()
    result_list = manager.list([None] * len(indices_mat))
    processes = []

    process_kwargs = {'csgraph':csgraph,'return_predecessors':return_predecessors,**kwargs}
    for i, origins in enumerate(indices_mat):
        process = Process(target=process_wrapper, args=(origins, process_kwargs, result_list, i))
        process.start()
        processes.append(process)
    for process in processes:
        process.join()
    return 0,0
    result_list = np.array(result_list,dtype="object")

    if return_predecessors == True: # result is a tuple
        dist_matrix = [res[0] for res in result_list]
        dist_matrix = np.concatenate(dist_matrix,axis=0)
        predecessors = [res[1] for res in result_list]
        predecessors = np.concatenate(predecessors,axis=0).astype(np.int32)
        return dist_matrix, predecessors
    else:
        dist_matrix = np.concatenate(result_list,axis=0)
        return dist_matrix
```

```python
from multiprocessing import Process, Manager, Queue
# Wrapper to split the indices (destination) into parallel batchs and compute the shortest path on each batchs.
def multiprocessing_dijkstra_Queue(csgraph,indices=None,return_predecessors=True, num_core=1,keep_running=True,**kwargs):
    '''
    num_core = 1 : number of threads.
    keep_running = False : if you want to keep the subprocesses alive (mmutiple dijktra in a loop for example)
    '''
    if num_core == 1:
        return  dijkstra(csgraph=csgraph, indices=indices, return_predecessors=return_predecessors,**kwargs)

    batch = round(len(indices)/num_core)
    indices_mat = [indices[i*batch:(1+i)*batch] for i in range(num_core-1)]
    indices_mat.append(indices[(num_core-1)*batch:])
    def process_wrapper(indices, kwargs, queue, index):
        result = dijkstra(indices=indices, **kwargs)
        queue.put(result)
        return result

    manager = Manager()
    queue = Queue()
    processes = []
    results = []

    process_kwargs = {'csgraph':csgraph,'return_predecessors':return_predecessors,**kwargs}
    for i, origins in enumerate(indices_mat):
        process = Process(target=process_wrapper, args=(origins, process_kwargs, queue, i))
        process.start()
        processes.append(process)
    for p in processes:
        results.append(queue.get())
    for process in processes:
        process.join()

    if return_predecessors == True: # result is a tuple
        dist_matrix = np.concatenate([res[0] for res in results],axis=0)
        predecessors = np.concatenate([res[1] for res in results],axis=0).astype(np.int32)
        return dist_matrix, predecessors
    else:
        dist_matrix = np.concatenate([res.result() for res in results],axis=0)
        return dist_matrix
```
